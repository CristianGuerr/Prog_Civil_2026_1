# Problema de Ingeniería (Nivel básico)

Un ingeniero agrícola necesita calcular **el área de un terreno rectangular** para estimar la cantidad de semillas necesarias.

La fórmula es:

[
Área = largo \times ancho
]

---

# 1. Definición del algoritmo (Entradas – Proceso – Salidas)

### Entradas

Datos que el programa debe recibir.

* `largo` del terreno (metros)
* `ancho` del terreno (metros)

### Proceso

Operación matemática a realizar.

* Calcular el área:

[
area = largo \times ancho
]

### Salidas

Resultado que mostrará el programa.

* Área del terreno en **metros cuadrados**

---

# 2. Diseño del Algoritmo

1. Leer el largo del terreno
2. Leer el ancho del terreno
3. Multiplicar largo por ancho
4. Mostrar el área calculada

---

# 3. Pseudocódigo

```
Inicio

    Leer largo
    Leer ancho

    area = largo * ancho

    Mostrar area

Fin
```

---

# 4. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica en ingeniería agrícola / civil

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "metros cuadrados")
```

### ¿Qué es Python?

Desde la perspectiva de un ingeniero, piensa en **Python** como una "navaja suiza" digital. Es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al inglés humano, facilitando que nos concentremos en la lógica del problema (el *qué* hacer) en lugar de complicaciones técnicas del hardware (el *cómo* lo entiende la máquina).

Para un estudiante de Ingeniería Civil, Agrícola o Agroindustrial, Python es ideal porque:

* **Versatilidad:** Sirve para analizar datos de sensores de humedad, calcular estructuras o crear aplicaciones web interactivas con **Streamlit**.
* **Comunidad:** Existen librerías específicas para casi todo (cálculo matricial, optimización, SIG, etc.).
* **Legibilidad:** Un código escrito hoy por ti será fácil de entender por un colega mañana.
