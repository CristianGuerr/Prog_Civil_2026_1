```python
# ENTRADAS: ancho y alto en metros
ancho = input("ingresen ancho de la ventana en metros")
ancho = float(ancho)
alto = input("ingresa el alto de la ventana en metros")
alto = float(alto)

# PROCESO: Calcula lm en pies y av en metros cuadrados
lm = 2 * (ancho + alto) * 3.25
av = (ancho * alto) * 2

# SALIDA: Sale la longitud de la madera (lm) en pies y área del vidrio en metros cuadrados
print("Longitud de la madera = ",lm,"pies de longitud")
print("Area de la ventana",av,"metros cuadrados")
```
